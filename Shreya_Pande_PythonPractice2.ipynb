# Problem 5. Country/Region with the Least Total Confirmed Cases (1 percentage point)

Find out which country/region has the least total confirmed cases. If a country/region has multiple provinces/states, the cases should be summed up for that specific country/region before comparing to other countries/regions

**Notes**
- "Total" means summing over confirmed cases for all days

```python
# Supply your code in this cell. Add more code cells as necessary.
# Add necessary comments to your code to explain your solution

confirmed_global_df
country_level_data_calc = confirmed_global_df.copy()
country_level_data_calc = country_level_data_calc.drop(['Province/State', 'Lat', 'Long'], axis=1) #Dropping extra columns

country_level_data_calc
country_level_data_calc['Country_Total']= country_level_data_calc.sum(axis=1) 
country_level_data_calc

country_level_data = country_level_data_calc.groupby(['Country_Region']).Country_Total.sum().reset_index(name="cummulative_country_count") 
country_level_data

country_level_data.sort_values(by = ['cummulative_country_count'], ascending = [True], na_position = 'first') #sorting data as required
```
